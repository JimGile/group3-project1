```python
# Import libraries
# source:
# https://databank.worldbank.org/source/food-prices-for-nutrition/Type/TABLE/preview/on#
# filters: 
# database - FPFN
# country - all
# series - all
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from matplotlib import pyplot as plt
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
csv_path = "../data/World_Bank/Food_Prices_For_Nutrition.csv"
food_nutrition_df = pd.read_csv(csv_path, encoding="ISO-8859-1")
food_nutrition_df.head()
```

```python
#Data
food_nutrition_df.count()
```

```python
#Data Cleaning
food_nutrition_df = food_nutrition_df.dropna(how="any")
food_nutrition_df.head()
```

```python
#After looking at the data it appears they give the most amount of data by the year of 2017 for every country which means we should narrow down our findings based on that. 
food_nutrition_2017_df = food_nutrition_df[food_nutrition_df["Time Code"]== "YR2017"]
food_nutrition_2017_df.head()
```

```python
# Write to CSV
food_nutrition_2017_df.to_csv("../data/World_Bank/food_nutrition_2017.csv")
```
