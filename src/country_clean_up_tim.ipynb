```python
# Extract relevant columns from each DataFrame
bank_gdp_selected = bank_gdp_df[['Country Name', 'Country Code'] + [str(year) for year in range(1960, 2024)]]
bank_food_selected = bank_food_df[['Country Name', 'Country Code']]
olympic_medals_selected = olympic_medals_df[['country_name', 'country_3_letter_code', 'medal_type']]
olympic_results_selected = olympic_results_df[['country_name', 'country_3_letter_code', 'medal_type']]
```

```python
# Reshape bank_gdp_df to have 'Year' as a column and 'GDP' as values
bank_gdp_long = pd.melt(bank_gdp_selected, id_vars=['Country Name', 'Country Code'], var_name='Year', value_name='GDP')
bank_gdp_long['Year'] = bank_gdp_long['Year'].astype(int)  # Ensure 'Year' is integer type
```

```python
# Create a common 'Year' column for the bank_food_df (assuming each row corresponds to a specific year)
# If 'Time' column represents the year in bank_food_df
bank_food_selected = bank_food_df[['Country Name', 'Country Code', 'Affordability of a healthy diet: ratio of cost to the food poverty line [CoHD_pov]','Time']].rename(columns={'Time': 'Year', 'Affordability of a healthy diet: ratio of cost to the food poverty line [CoHD_pov]': 'Healthy Diet Affordability'})
```

```python
# Handle non-finite values in 'Year'
bank_food_selected['Year'] = pd.to_numeric(bank_food_selected['Year'], errors='coerce')
bank_food_selected = bank_food_selected.dropna(subset=['Year'])
bank_food_selected['Year'] = bank_food_selected['Year'].astype(int)
```

```python
# Merging the DataFrames
merged_df = bank_gdp_long.merge(bank_food_selected, on=['Country Name', 'Country Code', 'Year'], how='left')
```

```python
# ChatGPT assistance for year
# Since olympic_medals_df and olympic_results_df do not contain a year column, 
# assume you want to join them on the country-related columns and medal type.
# This part is tricky as these DataFrames do not directly align with year data.
# If you have a separate logic to associate Olympic data with years, apply it accordingly.

# For simplicity, this example merges based on country and assumes the medal data is for all years combined.
olympic_combined = olympic_medals_selected.merge(olympic_results_selected, 
                                                 on=['country_name', 'country_3_letter_code', 'medal_type'], 
                                                 how='outer')
```

```python
# Final merge
final_df = merged_df.merge(olympic_combined, 
                           left_on=['Country Name', 'Country Code'], 
                           right_on=['country_name', 'country_3_letter_code'], 
                           how='left')

# Set 'Year' as the index
final_df.set_index('Year', inplace=True)
```

```python
# Drop duplicate columns from the merge
final_df.drop(columns=['country_name', 'country_3_letter_code'], inplace=True)
# Display the final DataFrame
# print(final_df.head())
```

```python
# Convert 'GDP' and 'Healthy Diet Affordability' columns to numeric types
final_df['GDP'] = pd.to_numeric(final_df['GDP'], errors='coerce')
final_df['Healthy Diet Affordability'] = pd.to_numeric(final_df['Healthy Diet Affordability'], errors='coerce')

# Calculate mean GDP and Healthy Diet Affordability by country
mean_values = final_df.groupby(['Country Name', 'Country Code']).agg({
    'GDP': 'mean',
    'Healthy Diet Affordability': 'mean'
}).reset_index()

print(mean_values)
```

```python
# Round GDP and Healthy Diet Affordability to 2 decimal places
mean_values['GDP'] = mean_values['GDP'].round(2)
mean_values['Healthy Diet Affordability'] = mean_values['Healthy Diet Affordability'].round(2)
print(mean_values)
```
